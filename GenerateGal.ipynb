```python
import h5py
import numpy as np
from astropy.table import Table,vstack
from astropy.io import ascii
```

```python
f = h5py.File('/global/homes/w/wballard/LDC2_sangria_training_v2.h5', 'r')
```

```python
#Create Detatched Binaries Table
dset=f['sky/dgb/cat']
dlen=len(dset)
detached_binaries=np.reshape(dset,len(dset))
dat_length=len(detached_binaries[0])
dgb_arr=np.empty((dlen,dat_length-1))
def tupToArr(x):
    temp=np.array(list(x))
    return temp[1:]

for i in range(dlen):
    dgb_arr[i]=tupToArr(detached_binaries[i])


dgb_arr=np.transpose(dgb_arr)
amp=dgb_arr[0]
ecLat=dgb_arr[1]
ecLong=dgb_arr[2]
freq=dgb_arr[3]
freqD=dgb_arr[4]
incl=dgb_arr[5]
initPhase=dgb_arr[6]
polz=dgb_arr[7]
dgb_table=Table([freq,freqD,ecLat,ecLong,amp,incl,initPhase,polz])
dgb_table
```

```python
#Create Interacting Binaries Table
dset=f['sky/igb/cat']
dlen=len(dset)
interacting_binaries=np.reshape(dset,len(dset))
dat_length=len(interacting_binaries[0])
igb_arr=np.empty((dlen,dat_length-1))

for i in range(dlen):
    igb_arr[i]=tupToArr(interacting_binaries[i])


igb_arr=np.transpose(igb_arr)
amp=igb_arr[0]
ecLat=igb_arr[1]
ecLong=igb_arr[2]
freq=igb_arr[3]
freqD=igb_arr[4]
incl=igb_arr[5]
initPhase=igb_arr[6]
polz=igb_arr[7]
igb_table=Table([freq,freqD,ecLat,ecLong,amp,incl,initPhase,polz])
#names=["freq","freqD","ecLat","ecLong","amp","incl","initPhase","polz"]
igb_table
```

```python
#Create verification binary tables
dset=f['sky/vgb/cat']
dlen=len(dset)
verif_binaries=np.reshape(dset,len(dset))
dat_length=len(verif_binaries[0])
vgb_arr=np.empty((dlen,dat_length-1))

for i in range(dlen):
    vgb_arr[i]=tupToArr(verif_binaries[i])


vgb_arr=np.transpose(vgb_arr)
amp=vgb_arr[0]
ecLat=vgb_arr[3]
ecLong=vgb_arr[4]
freq=vgb_arr[8]
freqD=vgb_arr[9]
incl=vgb_arr[12]
initPhase=vgb_arr[13]
polz=vgb_arr[17]
vgb_table=Table([freq,freqD,ecLat,ecLong,amp,incl,initPhase,polz])
vgb_table
```

```python
full_table=vstack([dgb_table,igb_table,vgb_table])
ascii.write(full_table,"full_galaxy.dat", format="no_header")
```

```python
#Add CVs
import pandas as pd

df=pd.read_csv("dat_maxDistance_1000_GW.txt",delimiter=" ")
CVs=Table.from_pandas(df)
full_tableCV=vstack([full_table,CVs])
full_tableCV
```

```python
ascii.write(full_tableCV, "full_galaxy_plus_CVs.dat",format="no_header")
```
